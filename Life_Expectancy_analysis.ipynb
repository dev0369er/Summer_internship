```python
healthexp=sns.load_dataset('healthexp')
print(healthexp.head())
```

```python
print(healthexp.info())
```

```python
print(healthexp.describe())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.scatterplot(x='Spending_USD',y='Life_Expectancy',hue='Country',data=healthexp)
plt.title('Scatter Plot of Spending vs Life Expectancy')
plt.xlabel('Spending USD')
plt.ylabel('Life Expectancy')
plt.legend(bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Country',y='Spending_USD',hue='Country',data=healthexp)
plt.title('Health Spending by Country')
plt.xlabel('Country')
plt.ylabel('Health Spending')
plt.xticks(rotation=80)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Country',y='Life_Expectancy',hue='Country',data=healthexp)
plt.title('Life Expectancy by Country')
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.swarmplot(x='Country',y='Life_Expectancy',hue='Country',data=healthexp)
plt.title('Life Expectancy by Country')
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Spending_USD',y='Life_Expectancy',hue='Country',data=healthexp,marker='o')
plt.title('Health Spending vs Life Expectancy')
plt.xlabel('Health Spending')
plt.ylabel('Life Expectancy')
plt.legend(bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.pointplot(x='Country',y='Spending_USD',hue='Country',data=healthexp)
plt.title('Life Spending by Country')
plt.xlabel('Country')
plt.ylabel('Health Spending')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.catplot(x='Country',y='Spending_USD',hue='Country',kind='bar',data=healthexp,height=6,aspect=2)
plt.title('Health Spending by Country')
plt.xlabel('Country')
plt.ylabel('Health Spending')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```
